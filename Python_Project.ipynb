Function #1

```python
import math
import re
def calculate_entropy(password):
  charset_size = 0
  if re.search(r"[a-z]", password):
    charset_size += 26
  if re.search(r"[A-Z]",password):
    charset_size += 26
  if re.search(r"[0-9]",password):
    charset_size +=10
  if re.search(r"[!@#$%^&*()_+\-=\[\]{};:'\",.<>/?\\|`~]",password):
    charset_size += 32
  if charset_size == 0:
    return 0
  entropy = len(password)*math.log2(charset_size)
  return entropy
```

```python
def evaluate_strength(entropy):
  if entropy < 28:
    return "Very Weak"
  elif entropy < 36:
    return "Weak"
  elif entropy < 60:
    return "Reasonable"
  elif entropy < 80:
    return "Strong"
  else:
    return "Very strong"
```
